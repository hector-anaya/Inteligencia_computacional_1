```python
import random
A=[]
B=[]
for i in range (2):
    A.append(input("Ingrese la zona A"))
for i in range(2):
    B.append(input("Ingrese la zona B"))
N=input("Numero de puntos a generar")
```

```python
A,B
```

```python
for i in range(2):
    A[i]=[int(A[i][0]),(int(A[i][2]))]
    B[i]=[int(B[i][0]),(int(B[i][2]))]
```

```python
N=int(N)
```

```python
index_values = [1, 2, 3, 4]
column_values = ['X', 'Y']
dfA = pd.DataFrame(data = A,
                  columns = column_values)
dfB = pd.DataFrame(data = B,
                  columns = column_values)
```

```python
dfA,dfB
```

```python
PA=[]
PB=[]
cc=2
dd=2
while cc-N<N:
    x=random.randint(min(A)[0],max(A)[0])
    y=random.randint(min(A)[1],max(A)[1])

    c=0

    for j in range(cc):
        if x==A[j][0] and y==A[j][1]:        
            c=c+1


    if c<=0:
        PA.append([x,y])
        A.append([x,y])
        cc=cc+1

while dd-N<N:
    x2=random.randint(min(B)[0],max(B)[0])
    y2=random.randint(min(B)[1],max(B)[1])    
    d=0
    for j in range(dd):
        if x2==B[j][0] and y2==B[j][1]:  
            d=d+1
    if d<=0:
        PB.append([x2,y2])
        B.append([x2,y2])
        dd=dd+1    
```

```python
PA
```

```python
df2 = pd.DataFrame(data = PA, 
                  columns = column_values)
df3=pd.DataFrame(data = PB,  
                  columns = column_values)
```

```python
df2,df3
```

```python
df=pd.concat([dfA,dfB,df2,df3])
```
